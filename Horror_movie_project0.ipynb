```python
df3= df.sort_values(by='release_date')
df3 = df3.groupby(df.release_date.dt.year).popularity.mean()
plt.figure(figsize=(12,6))
sns.lineplot(data=df3).set_xticks(range(df3.index[0],df3.index[-1] + 1,4))
plt.show()
```

```python
avg_votes = df[['title','vote_average']].sort_values(by='vote_average',ascending=False)
avg_votes
```

```python
import numpy as np
```

```python
df["vote_avg_ratio"] = df['vote_average']/df["vote_count"]
df['vote_avg_ratio'] = df["vote_avg_ratio"].replace({np.inf:0,np.nan:0})
df[['title','vote_avg_ratio']].sort_values(by='vote_avg_ratio',ascending=False)
```

```python
sub_df = df[['popularity','title','budget','vote_average']]
sub_df = sub_df[sub_df.budget != 0]
sub_df = sub_df.sort_values(by='popularity',ascending=False).head(20)
sub_df
```

```python
ax1 = sns.set_style(style=None, rc=None )
fig, ax1 = plt.subplots(figsize=(20,6))
plt.xticks(rotation=45)
plt.title("Popularity x budget for each movie")
sns.lineplot(data = sub_df['popularity'], marker='o', sort = False, ax=ax1)
ax2 = ax1.twinx()
sns.barplot(data=sub_df,x = sub_df.title,y = sub_df.budget,ax=ax2,alpha=0.5,color='orange')
plt.show()
```
